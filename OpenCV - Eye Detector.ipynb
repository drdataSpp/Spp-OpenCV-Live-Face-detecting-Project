## Creating the Eye Detector Variable

```python
eye_detector = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
```

## Creating a function to detect eyes

```python
def detect_eyes(image):
    img = image.copy()
    points = eye_detector.detectMultiScale(img, scaleFactor=1.2, minNeighbors=5)
    
    #Grabbing the points from the above variable to plot the face in the image
    for (x,y,w,h) in points:
        cv2.rectangle(img, (x,y), (x+w, y+h), (255,255,255,), 10)
        
    #Here we draw white rectangle of thickness 10 around the detected place.
    #Top Left of the rectangle = x,y
    #Bottom right of the rectangle = x+w, y+h
    
    return img
```

## Detecting the Model

```python
Detect_Tarantino_eye = detect_eyes(tarantino)
plt.imshow(Detect_Tarantino_eye, cmap='gray')
```

```python
Detect_Angelina_eye = detect_eyes(angelina)
plt.imshow(Detect_Angelina_eye, cmap='gray')
```

```python
Detect_beans_eye = detect_eyes(Mr_Bean)
plt.imshow(Detect_beans_eye, cmap='gray')
```

## Detecting Faces In Real-Time using a Webcam

```python
webcam = cv2.VideoCapture(0)

while True:
    ret, frame = webcam.read(0)
    
    frame = detect_eyes(frame)
    
    cv2.imshow("Your eyes is being detected", frame)
    
    exitKey = cv2.waitKey(10)
    if exitKey == 27:
        break
        
webcam.release()
```
