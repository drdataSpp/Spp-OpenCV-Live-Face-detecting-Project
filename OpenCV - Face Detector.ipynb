## Creating the face detector variable 

```python
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

## Creating a function to detect faces 

```python
def detect_face(image):
    img = image.copy()
    points = face_detector.detectMultiScale(img, scaleFactor=1.2, minNeighbors=5)
    
    #Grabbing the points from the above variable to plot the face in the image
    for (x,y,w,h) in points:
        cv2.rectangle(img, (x,y), (x+w, y+h), (255,255,255,), 10)
        
    #Here we draw white rectangle of thickness 10 around the detected place.
    #Top Left of the rectangle = x,y
    #Bottom right of the rectangle = x+w, y+h
    
    return img
```

### Detecting Tarantino's Face 

```python
Detect_Tarantino_face = detect_face(tarantino)
```

```python
plt.imshow(Detect_Tarantino_face, cmap='gray')
```

#### Result: The Face recognizer model worked well and it has also detected the face-like structures around the main subject.

### Detecting Angelina's Face

```python
Detect_Angelina_face = detect_face(angelina)
```

```python
plt.imshow(Detect_Angelina_face, cmap='gray')
```

#### Result: The Face recognizer model worked well.

### Detecting the faces in the Argentina Football Team

```python
Detect_Team = detect_face(argentina_team)
```

```python
plt.imshow(Detect_Team, cmap='gray')
```

#### Result: The Face recognizer model worked well in detecting all the faces in the team but did 2 incorrect detections.

## Detecting Faces In Real-Time using a Webcam

```python
webcam = cv2.VideoCapture(0)

while True:
    ret, frame = webcam.read(0)
    
    frame = detect_face(frame)
    
    cv2.imshow("Your face is being detected", frame)
    
    exitKey = cv2.waitKey(10)
    if exitKey == 27:
        break
        
webcam.release()
```
